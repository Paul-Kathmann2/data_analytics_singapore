```python
#importing libraries
import pandas as pd
import numpy as np
import quandl
import matplotlib.pyplot as plt
```

```python
#quandl data sets
ppp = quandl.get("ODA/SGP_PPPEX", authtoken="cPaLk-kXXEd6ZNsAp8fy", start_date="1980-12-31", end_date="2018-12-31")
er = quandl.get("FED/RXI_N_A_SI", authtoken="cPaLk-kXXEd6ZNsAp8fy", start_date="1980-12-31", end_date="2018-12-31")
gross_savings = quandl.get("ODA/SGP_NGSD_NGDP", authtoken="cPaLk-kXXEd6ZNsAp8fy", start_date="1980-12-31", end_date="2018-12-31")
current_account_balance = quandl.get("ODA/SGP_BCA", authtoken="cPaLk-kXXEd6ZNsAp8fy", start_date="1980-12-31", end_date="2018-12-31")
current_account_percentage_gdp = quandl.get("ODA/SGP_BCA_NGDPD", authtoken="cPaLk-kXXEd6ZNsAp8fy", start_date="1980-12-31", end_date="2018-12-31")

#converting the data sets to pandas DataFrames
ppp_er = pd.DataFrame(ppp)
actual_er = pd.DataFrame(er)
grs_savings = pd.DataFrame(gross_savings)
current_act = pd.DataFrame(current_account_balance)
current_act_percent = pd.DataFrame(current_account_percentage_gdp)
```

```python
#convert into SGD

ppp_er_sgd = 1 / ppp_er
actual_er_sgd = 1 / actual_er

#print(ppp_er_sgd.head())
#print(actual_er_sgd.head())

#plot the inverted exchange rate
plt.plot(ppp_er_sgd, label="PPP implied exchange rate")
plt.xlabel("year")
plt.ylabel("PPP implied ER, LCU per USD")

# actual exchange rate
#plt.subplot(1, 2, 2)
plt.plot(actual_er_sgd, label="actual exchange rate")
plt.xlabel("year")
plt.ylabel("USD/SGD")
plt.tight_layout()
plt.legend(loc="upper right")
plt.grid()
plt.show()
```
